```python
import cv2
import numpy as np
```

```python

# Load the satellite image
img = cv2.imread('/content/campus_jpeg.jpeg')

# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply a Laplacian filter to enhance edges
laplacian = cv2.Laplacian(img, cv2.CV_64F)

# Adjust the contrast of the image using histogram equalization
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
equ = cv2.equalizeHist(gray)

# Apply Gaussian smoothing to remove noise
blur = cv2.GaussianBlur(equ,(5,5),0)

# Define the High Pass filter kernel
kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])

# Apply the High Pass filter to sharpen the image
sharp = cv2.filter2D(gray, -1, kernel, borderType=cv2.BORDER_CONSTANT)

# Combine the original and processed images horizontally
combined = np.hstack([gray, sharp])

# Display the combined image
cv2_imshow(combined)
```

```python
# Save the processed image to your Google Drive
cv2.imwrite('/content/satellite_image_gray.jpg', combined)
```
